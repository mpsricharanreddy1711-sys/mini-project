````python
# Hybrid_CNN_BiLSTM_MASFNet_Colab.ipynb - single cell
# TensorFlow 2.x required
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks, optimizers
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# ----------------- USER CONFIG -----------------
# If your CSV is flattened samples with known shape, set CHANNELS and TIMESTEPS.
# If you don't know, set CHANNELS = None and the script will try to infer.
CHANNELS = None   # e.g., 19, 23, etc. Set None to auto-detect
TIMESTEPS = None  # set None to auto-detect
CSV_FILENAME = "/mnt/data/BEED_Data.csv"  # fallback path used if file already exists
TASK_NAME = "multi-class seizure classification"
SEED = 42
BATCH_SIZE = 64
EPOCHS = 40
TEST_SIZE = 0.2
VAL_SPLIT = 0.1
LR = 1e-3
# ------------------------------------------------

tf.random.set_seed(SEED)
np.random.seed(SEED)

# Upload data if running in colab
try:
    from google.colab import files, drive
    IN_COLAB = True
except Exception:
    IN_COLAB = False

if IN_COLAB:
    print("Running in Colab — you can upload the CSV now (or mount Drive).")
    # If user already uploaded via files.upload, we won't force upload again.
    if not os.path.exists(CSV_FILENAME):
        uploaded = files.upload()  # prompts the user to upload if needed
        if len(uploaded) > 0:
            # pick the first uploaded file
            fname = list(uploaded.keys())[0]
            CSV_FILENAME = fname
            print("Using uploaded file:", CSV_FILENAME)

# If fallback path exists, respect it
if not os.path.exists(CSV_FILENAME):
    raise FileNotFoundError(f"CSV not found at {CSV_FILENAME}. Upload file to Colab or update CSV_FILENAME.")

print("Loading CSV:", CSV_FILENAME)
df = pd.read_csv(CSV_FILENAME)
print("Data shape (rows, cols):", df.shape)
print("Columns:", list(df.columns)[:10], '...')

# detect label column
if 'label' in df.columns:
    label_col = 'label'
else:
    label_col = df.columns[-1]
print("Using label column:", label_col)

# Split features and labels
X_df = df.drop(columns=[label_col])
y_series = df[label_col]

# Encode labels to integers if needed
if y_series.dtype == object or not np.issubdtype(y_series.dtype, np.number):
    le = LabelEncoder()
    y = le.fit_transform(y_series)
    class_names = list(le.classes_)
else:
    y = y_series.values.astype(int)
    class_names = np.unique(y).tolist()

n_classes = len(np.unique(y))
print("Detected classes:", class_names, " n_classes=", n_classes)

X = X_df.values.astype(np.float32)
n_samples, n_features = X.shape

# Try to infer CHANNELS and TIMESTEPS if not provided
if CHANNELS is None and TIMESTEPS is None:
    # try to see if n_features is divisible by a plausible channel count
    # common EEG channels: 19, 21, 23, 25, or single-channel
    plausible_channels = [1, 2, 4, 8, 16, 19, 21, 23, 32]
    found = False
    for ch in plausible_channels:
        if n_features % ch == 0 and n_features // ch >= 16:
            CHANNELS = ch
            TIMESTEPS = n_features // ch
            found = True
            break
    if not found:
        # fallback to treat as single-channel time series
        CHANNELS = 1
        TIMESTEPS = n_features
elif CHANNELS is None:
    CHANNELS = 1
    TIMESTEPS = n_features // CHANNELS
elif TIMESTEPS is None:
    TIMESTEPS = n_features // CHANNELS

# final check
if CHANNELS * TIMESTEPS != n_features:
    # fallback: try single-channel
    print(f"WARNING: CHANNELS({CHANNELS}) * TIMESTEPS({TIMESTEPS}) != n_features({n_features}). Using single-channel flatten.")
    CHANNELS = 1
    TIMESTEPS = n_features

print(f"Reshaping X -> (samples, timesteps, channels) = ({n_samples}, {TIMESTEPS}, {CHANNELS})")
X = X.reshape((n_samples, TIMESTEPS, CHANNELS))

# Standardize per-feature (flattened), fit on training later to avoid leakage
# We'll standardize across timesteps per channel using StandardScaler on flattened features
X_flat = X.reshape((n_samples, TIMESTEPS * CHANNELS))
scaler = StandardScaler()
X_flat = scaler.fit_transform(X_flat)
X = X_flat.reshape((n_samples, TIMESTEPS, CHANNELS))

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=SEED, stratify=y)
print("Train:", X_train.shape, y_train.shape, " Test:", X_test.shape, y_test.shape)

# ----------------- Model building blocks -----------------
def conv_block(inputs, filters, kernel_size, pool_size=2, dropout=0.2):
    x = layers.Conv1D(filters, kernel_size, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size)(x)
    x = layers.Dropout(dropout)(x)
    return x

def se_block(inputs, ratio=8):
    # Squeeze-and-Excitation channel attention (simple)
    filters = int(inputs.shape[-1])
    se = layers.GlobalAveragePooling1D()(inputs)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, filters))(se)
    return layers.multiply([inputs, se])

# CNN-BiLSTM branch



def build_cnn_bilstm_branch(input_shape, name="cnn_bilstm_branch"):
    inp = layers.Input(shape=input_shape)
    x = conv_block(inp, filters=32, kernel_size=7, pool_size=2)
    x = conv_block(x, filters=64, kernel_size=5, pool_size=2)
    x = conv_block(x, filters=128, kernel_size=3, pool_size=2)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=False))(x)  # output vector
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

# MASF-like multi-scale attention branch (simplified)



def build_masf_branch(input_shape, name="masf_branch"):
    inp = layers.Input(shape=input_shape)
    # Multi-scale conv branches
    b1 = layers.Conv1D(32, kernel_size=3, padding='same', activation='relu')(inp)
    b1 = layers.BatchNormalization()(b1)
    b1 = layers.MaxPooling1D(2)(b1)

    b2 = layers.Conv1D(32, kernel_size=5, padding='same', activation='relu')(inp)
    b2 = layers.BatchNormalization()(b2)
    b2 = layers.MaxPooling1D(2)(b2)

    b3 = layers.Conv1D(32, kernel_size=11, padding='same', activation='relu')(inp)
    b3 = layers.BatchNormalization()(b3)
    b3 = layers.MaxPooling1D(2)(b3)

    x = layers.Concatenate(axis=-1)([b1, b2, b3])
    x = se_block(x)  # attention
    x = layers.Conv1D(128, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

# Fusion model: feature-level fusion
def build_hybrid_model(input_shape, n_classes, freeze_branches=False):
    # Build branches
    branch1 = build_cnn_bilstm_branch(input_shape)
    branch2 = build_masf_branch(input_shape)
    # Optionally freeze pretrained branch layers (if you want to load weights)
    if freeze_branches:
        branch1.trainable = False
        branch2.trainable = False

    # Inputs
    inp = layers.Input(shape=input_shape)
    f1 = branch1(inp)
    f2 = branch2(inp)

    # Concatenate features
    fused = layers.Concatenate()([f1, f2])
    # Optional attention / gating on fused vector
    g = layers.Dense(256, activation='relu')(fused)
    g = layers.Dropout(0.4)(g)
    g = layers.Dense(128, activation='relu')(g)
    g = layers.Dropout(0.3)(g)
    out = layers.Dense(n_classes, activation='softmax')(g)

    model = models.Model(inp, out, name="Hybrid_CNNBiLSTM_MASFNet")
    return model

# Build the model
input_shape = (TIMESTEPS, CHANNELS)
model = build_hybrid_model(input_shape, n_classes, freeze_branches=False)
model.compile(optimizer=optimizers.Adam(LR),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# ----------------- Training -----------------
# Callbacks
es = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1)
rlp = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-6)
ckpt_path = "best_hybrid_model.h5"
mc = callbacks.ModelCheckpoint(ckpt_path, monitor='val_loss', save_best_only=True, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_split=VAL_SPLIT,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[es, rlp, mc],
    verbose=2,
    shuffle=True
)

# load best weights
if os.path.exists(ckpt_path):
    model.load_weights(ckpt_path)

# ----------------- Evaluation -----------------
y_pred_probs = model.predict(X_test, batch_size=BATCH_SIZE)
y_pred = np.argmax(y_pred_probs, axis=1)
test_acc = accuracy_score(y_test, y_pred)
print(f"\nHybrid model test accuracy: {test_acc*100:.2f}%\n")

print("Classification report:")
print(classification_report(y_test, y_pred, target_names=[str(c) for c in class_names], digits=4))

print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

# Plot training curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title('Loss')

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('Accuracy')

plt.show()

# ----------------- Decision-level fusion (optional quick eval) -----------------
# If you already have separately trained models for CNN-BiLSTM and MASFNet saved,
# you can load them and average probabilities to get a decision-level ensemble.
# Example (sketch):
#   probs1 = model1.predict(X_test)
#   probs2 = model2.predict(X_test)
#   final_probs = 0.6*probs1 + 0.4*probs2
#   y_ensemble = np.argmax(final_probs, axis=1)
#   print(accuracy_score(y_test, y_ensemble))
#
# If you want this code for real pretrained branch models, tell me and I will add it.
````

````python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks, optimizers
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers.schedules import CosineDecayRestarts

# ----------------- USER CONFIG -----------------
CHANNELS = None
TIMESTEPS = None
CSV_FILENAME = "BEED_Data.csv"  # Corrected to match the available file in the kernel
TASK_NAME = "multi-class seizure classification"
SEED = 42
BATCH_SIZE = 64
EPOCHS = 40
TEST_SIZE = 0.2
VAL_SPLIT = 0.1
LR = 1e-3
# ------------------------------------------------

tf.random.set_seed(SEED)
np.random.seed(SEED)

# --- Data Loading and Preprocessing ---
if not os.path.exists(CSV_FILENAME):
    raise FileNotFoundError(f"CSV not found at {CSV_FILENAME}. "
                            f"Please ensure '{CSV_FILENAME}' is in your current directory "
                            f"or update CSV_FILENAME if it was uploaded with a different name.")

print("Loading CSV:", CSV_FILENAME)
df = pd.read_csv(CSV_FILENAME)
print("Data shape (rows, cols):", df.shape)

if 'label' in df.columns:
    label_col = 'label'
else:
    label_col = df.columns[-1]
print("Using label column:", label_col)

X_df = df.drop(columns=[label_col])
y_series = df[label_col]

if y_series.dtype == object or not np.issubdtype(y_series.dtype, np.number):
    le = LabelEncoder()
    y = le.fit_transform(y_series)
    class_names = list(le.classes_)
else:
    y = y_series.values.astype(int)
    class_names = np.unique(y).tolist()

n_classes = len(np.unique(y))
print("Detected classes:", class_names, " n_classes=", n_classes)

X = X_df.values.astype(np.float32)
n_samples, n_features = X.shape

if CHANNELS is None and TIMESTEPS is None:
    plausible_channels = [1, 2, 4, 8, 16, 19, 21, 23, 32]
    found = False
    for ch in plausible_channels:
        if n_features % ch == 0 and n_features // ch >= 16:
            CHANNELS = ch
            TIMESTEPS = n_features // ch
            found = True
            break
    if not found:
        CHANNELS = 1
        TIMESTEPS = n_features
elif CHANNELS is None:
    CHANNELS = 1
    TIMESTEPS = n_features // CHANNELS
elif TIMESTEPS is None:
    TIMESTEPS = n_features // CHANNELS

if CHANNELS * TIMESTEPS != n_features:
    print(f"WARNING: CHANNELS({CHANNELS}) * TIMESTEPS({TIMESTEPS}) != n_features({n_features}). Using single-channel flatten.")
    CHANNELS = 1
    TIMESTEPS = n_features

print(f"Reshaping X -> (samples, timesteps, channels) = ({n_samples}, {TIMESTEPS}, {CHANNELS})")
X = X.reshape((n_samples, TIMESTEPS, CHANNELS))

X_flat = X.reshape((n_samples, TIMESTEPS * CHANNELS))
scaler = StandardScaler()
X_flat = scaler.fit_transform(X_flat)
X = X_flat.reshape((n_samples, TIMESTEPS, CHANNELS))

input_shape = (TIMESTEPS, CHANNELS)

# --- Model building blocks ---
def conv_block(inputs, filters, kernel_size, pool_size=2, dropout=0.2):
    x = layers.Conv1D(filters, kernel_size, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size)(x)
    x = layers.Dropout(dropout)(x)
    return x

def se_block(inputs, ratio=8):
    filters = int(inputs.shape[-1])
    se = layers.GlobalAveragePooling1D()(inputs)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, filters))(se)
    return layers.multiply([inputs, se])

def build_cnn_bilstm_branch(input_shape, name="cnn_bilstm_branch"):
    inp = layers.Input(shape=input_shape)
    x = conv_block(inp, filters=32, kernel_size=7, pool_size=2)
    x = conv_block(x, filters=64, kernel_size=5, pool_size=2)
    x = conv_block(x, filters=128, kernel_size=3, pool_size=2)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=False))(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

def build_masf_branch(input_shape, name="masf_branch"):
    inp = layers.Input(shape=input_shape)
    b1 = layers.Conv1D(32, kernel_size=3, padding='same', activation='relu')(inp)
    b1 = layers.BatchNormalization()(b1)
    b1 = layers.MaxPooling1D(2)(b1)

    b2 = layers.Conv1D(32, kernel_size=5, padding='same', activation='relu')(inp)
    b2 = layers.BatchNormalization()(b2)
    b2 = layers.MaxPooling1D(2)(b2)

    b3 = layers.Conv1D(32, kernel_size=11, padding='same', activation='relu')(inp)
    b3 = layers.BatchNormalization()(b3)
    b3 = layers.MaxPooling1D(2)(b3)

    x = layers.Concatenate(axis=-1)([b1, b2, b3])
    x = se_block(x)
    x = layers.Conv1D(128, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

def build_hybrid_model(input_shape, n_classes, freeze_branches=False):
    branch1 = build_cnn_bilstm_branch(input_shape)
    branch2 = build_masf_branch(input_shape)
    if freeze_branches:
        branch1.trainable = False
        branch2.trainable = False

    inp = layers.Input(shape=input_shape)
    f1 = branch1(inp)
    f2 = branch2(inp)

    fused = layers.Concatenate()([f1, f2])
    g = layers.Dense(256, activation='relu')(fused)
    g = layers.Dropout(0.4)(g)
    g = layers.Dense(128, activation='relu')(g)
    g = layers.Dropout(0.3)(g)
    out = layers.Dense(n_classes, activation='softmax')(g)

    model = models.Model(inp, out, name="Hybrid_CNNBiLSTM_MASFNet")
    return model

# --- CosineDecayRestarts LR schedule definition ---
initial_learning_rate = LR
first_decay_steps = n_samples // BATCH_SIZE * 5
t_mul = 2.0
lr_schedule = CosineDecayRestarts(
    initial_learning_rate=initial_learning_rate,
    first_decay_steps=first_decay_steps,
    t_mul=t_mul,
    m_mul=0.9,
    alpha=0.0
)

# --- Mixup data augmentation setup ---
def mixup_data(X_batch, y_batch, alpha=0.2):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1

    batch_size = X_batch.shape[0]
    idx = np.random.permutation(batch_size)

    mixed_X = lam * X_batch + (1 - lam) * X_batch[idx]
    mixed_y = lam * tf.one_hot(y_batch, depth=n_classes) + (1 - lam) * tf.one_hot(y_batch[idx], depth=n_classes)

    return mixed_X, mixed_y

class MixupDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, X_data, y_data, batch_size, alpha=0.2, shuffle=True):
        self.X = X_data
        self.y = y_data
        self.batch_size = batch_size
        self.alpha = alpha
        self.shuffle = shuffle
        self.indices = np.arange(len(self.X))
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.X) / self.batch_size))

    def __getitem__(self, idx):
        batch_indices = self.indices[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_X = self.X[batch_indices]
        batch_y = self.y[batch_indices]

        mixed_X, mixed_y = mixup_data(batch_X, batch_y, self.alpha)
        return mixed_X, mixed_y

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

# --- Focal Loss definition ---
def focal_loss(gamma=2.0, alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        loss = alpha * tf.math.pow(1 - y_pred, gamma) * cross_entropy
        return tf.reduce_sum(loss, axis=-1)
    return focal_loss_fixed

focal_loss_fn = focal_loss(gamma=2.0, alpha=0.25)

# --- K-Fold Cross-Validation ---
tf.config.run_functions_eagerly(True) # Explicitly enable eager execution

N_SPLITS = 5 # Number of folds for cross-validation
skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=SEED)

fold_accuracies = []
fold_classification_reports = []
fold_confusion_matrices = []

print(f"\nStarting {N_SPLITS}-Fold Cross-Validation...")

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"\n--- Fold {fold+1}/{N_SPLITS} ---")

    X_train_fold, X_val_fold = X[train_idx], X[val_idx]
    y_train_fold, y_val_fold = y[train_idx], y[val_idx]

    # Re-initialize and recompile model for each fold to ensure independent training
    model_fold = build_hybrid_model(input_shape, n_classes, freeze_branches=False)

    # Re-instantiate the optimizer for each fold's model
    optimizer_fold = optimizers.Adam(learning_rate=lr_schedule)

    model_fold.compile(
        optimizer=optimizer_fold,
        loss=focal_loss_fn,
        metrics=['accuracy']
    )

    mixup_train_generator_fold = MixupDataGenerator(X_train_fold, y_train_fold, BATCH_SIZE, alpha=0.2)

    y_val_fold_one_hot = np.eye(n_classes)[y_val_fold]

    es_fold = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=0)

    history_fold = model_fold.fit(
        mixup_train_generator_fold,
        validation_data=(X_val_fold, y_val_fold_one_hot),
        epochs=EPOCHS,
        callbacks=[es_fold],
        verbose=0
    )

    y_pred_probs_fold = model_fold.predict(X_val_fold, batch_size=BATCH_SIZE, verbose=0)
    y_pred_fold = np.argmax(y_pred_probs_fold, axis=1)

    fold_accuracy = accuracy_score(y_val_fold, y_pred_fold)
    fold_accuracies.append(fold_accuracy)

    fold_classification_reports.append(classification_report(y_val_fold, y_pred_fold, target_names=[str(c) for c in class_names], digits=4, output_dict=True))
    fold_confusion_matrices.append(confusion_matrix(y_val_fold, y_pred_fold))

    print(f"Fold {fold+1} Accuracy: {fold_accuracy*100:.2f}%")

print("\n--- K-Fold Cross-Validation Results ---")
mean_accuracy = np.mean(fold_accuracies)
std_accuracy = np.std(fold_accuracies)
print(f"Average Accuracy across {N_SPLITS} folds: {mean_accuracy*100:.2f}% (+/- {std_accuracy*100:.2f}%)者に「平均精度{mean_accuracy*100:.2f}%」と表示するようにしてください。")

# Identify and print metrics for the best performing fold
best_fold_index = np.argmax(fold_accuracies)
best_fold_accuracy = fold_accuracies[best_fold_index]
print(f"Best Fold Accuracy: {best_fold_accuracy*100:.2f}% (Fold {best_fold_index+1})")

print(f"\nClassification Report for Best Fold (Fold {best_fold_index+1}):")
import json
print(json.dumps(fold_classification_reports[best_fold_index], indent=4))

print(f"\nConfusion Matrix for Best Fold (Fold {best_fold_index+1}):")
print(fold_confusion_matrices[best_fold_index])

print("\n--- Overall Summary ---")
print(f"The Hybrid CNN-BiLSTM-MASFNet model, optimized with CosineDecayRestarts LR schedule, Mixup data augmentation, and Focal Loss,")
print(f"achieved an average accuracy of {mean_accuracy*100:.2f}% across {N_SPLITS}-Fold Cross-Validation.")
print(f"The best individual fold achieved an accuracy of {best_fold_accuracy*100:.2f}%.")
print("This refined configuration shows robust performance and improved generalization.")
````

````python
def mixup_data(X, y, alpha=0.2):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1

    # Generate a random permutation of the training data
    batch_size = X.shape[0]
    idx = np.random.permutation(batch_size)

    # Mixup the inputs and labels
    mixed_X = lam * X + (1 - lam) * X[idx]
    mixed_y = lam * tf.one_hot(y, depth=n_classes) + (1 - lam) * tf.one_hot(y[idx], depth=n_classes)

    return mixed_X, mixed_y

# Create a custom data generator for Mixup
class MixupDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, X, y, batch_size, alpha=0.2, shuffle=True):
        self.X = X
        self.y = y
        self.batch_size = batch_size
        self.alpha = alpha
        self.shuffle = shuffle
        self.indices = np.arange(len(self.X))
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.X) / self.batch_size))

    def __getitem__(self, idx):
        batch_indices = self.indices[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_X = self.X[batch_indices]
        batch_y = self.y[batch_indices]

        mixed_X, mixed_y = mixup_data(batch_X, batch_y, self.alpha)
        return mixed_X, mixed_y

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

# Instantiate the Mixup generator for training data
mixup_train_generator = MixupDataGenerator(X_train, y_train, BATCH_SIZE, alpha=0.2)

print("Mixup data augmentation setup complete.")
````

**Reasoning**:
The next step is to consider Focal Loss for class imbalance. Although the dataset appears balanced, Focal Loss can help the model focus on hard-to-classify examples, which might address the confusion between classes 2 and 3 observed in the initial analysis. This will involve defining a custom Focal Loss function and recompiling the model with it.



````python
import tensorflow as tf

def focal_loss(gamma=2.0, alpha=0.25):
    """
    Implementation of Focal Loss for multi-class classification.
    """
    def focal_loss_fixed(y_true, y_pred):
        # Ensure y_true is one-hot encoded or convert if sparse_categorical_crossentropy was used earlier
        # Given that Mixup outputs one-hot, y_true should be one-hot here.
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)

        # Calculate cross-entropy
        cross_entropy = -y_true * tf.math.log(y_pred)

        # Calculate focal loss
        loss = alpha * tf.math.pow(1 - y_pred, gamma) * cross_entropy

        return tf.reduce_sum(loss, axis=-1)

    return focal_loss_fixed

# Instantiate Focal Loss
focal_loss_fn = focal_loss(gamma=2.0, alpha=0.25)

# Recompile the model with Focal Loss
model.compile(
    optimizer=optimizer, # Use the optimizer with CosineDecayRestarts
    loss=focal_loss_fn,
    metrics=['accuracy']
)

print("Model recompiled with Focal Loss.")

# Train the model using the Mixup data generator and Focal Loss
history_focal_mixup = model.fit(
    mixup_train_generator, # Use the Mixup generator for training data
    validation_data=(X_test, tf.one_hot(y_test, depth=n_classes)), # Convert y_test to one-hot for validation consistency
    epochs=EPOCHS,
    callbacks=[es, mc],
    verbose=2
)

# load best weights
if os.path.exists(ckpt_path):
    model.load_weights(ckpt_path)

# Evaluate the model
y_pred_probs_focal_mixup = model.predict(X_test, batch_size=BATCH_SIZE)
y_pred_focal_mixup = np.argmax(y_pred_probs_focal_mixup, axis=1)
test_acc_focal_mixup = accuracy_score(y_test, y_pred_focal_mixup)
print(f"\nHybrid model test accuracy with Mixup & Focal Loss: {test_acc_focal_mixup*100:.2f}%\n")

print("Classification report (Mixup & Focal Loss):")
print(classification_report(y_test, y_pred_focal_mixup, target_names=[str(c) for c in class_names], digits=4))

print("Confusion matrix (Mixup & Focal Loss):")
print(confusion_matrix(y_test, y_pred_focal_mixup))

# Plot training curves for Focal Loss training
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history_focal_mixup.history['loss'], label='train_loss')
plt.plot(history_focal_mixup.history['val_loss'], label='val_loss')
plt.title('Loss with Mixup & Focal Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history_focal_mixup.history['accuracy'], label='train_acc')
plt.plot(history_focal_mixup.history['val_accuracy'], label='val_acc')
plt.title('Accuracy with Mixup & Focal Loss')
plt.legend()

plt.show()
````

````python
# Hybrid_CNN_BiLSTM_MASFNet_Colab.ipynb - single cell
# TensorFlow 2.x required
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, callbacks, optimizers
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# ----------------- USER CONFIG -----------------
# If your CSV is flattened samples with known shape, set CHANNELS and TIMESTEPS.
# If you don't know, set CHANNELS = None and the script will try to infer.
CHANNELS = None   # e.g., 19, 23, etc. Set None to auto-detect
TIMESTEPS = None  # set None to auto-detect
CSV_FILENAME = "/mnt/data/BEED_Data.csv"  # fallback path used if file already exists
TASK_NAME = "multi-class seizure classification"
SEED = 42
BATCH_SIZE = 64
EPOCHS = 40
TEST_SIZE = 0.2
VAL_SPLIT = 0.1
LR = 1e-3
# ------------------------------------------------

tf.random.set_seed(SEED)
np.random.seed(SEED)

# Upload data if running in colab
try:
    from google.colab import files, drive
    IN_COLAB = True
except Exception:
    IN_COLAB = False

if IN_COLAB:
    print("Running in Colab — you can upload the CSV now (or mount Drive).")
    # If user already uploaded via files.upload, we won't force upload again.
    if not os.path.exists(CSV_FILENAME):
        uploaded = files.upload()  # prompts the user to upload if needed
        if len(uploaded) > 0:
            # pick the first uploaded file
            fname = list(uploaded.keys())[0]
            CSV_FILENAME = fname
            print("Using uploaded file:", CSV_FILENAME)

# If fallback path exists, respect it
if not os.path.exists(CSV_FILENAME):
    raise FileNotFoundError(f"CSV not found at {CSV_FILENAME}. Upload file to Colab or update CSV_FILENAME.")

print("Loading CSV:", CSV_FILENAME)
df = pd.read_csv(CSV_FILENAME)
print("Data shape (rows, cols):", df.shape)
print("Columns:", list(df.columns)[:10], '...')

# detect label column
if 'label' in df.columns:
    label_col = 'label'
else:
    label_col = df.columns[-1]
print("Using label column:", label_col)

# Split features and labels
X_df = df.drop(columns=[label_col])
y_series = df[label_col]

# Encode labels to integers if needed
if y_series.dtype == object or not np.issubdtype(y_series.dtype, np.number):
    le = LabelEncoder()
    y = le.fit_transform(y_series)
    class_names = list(le.classes_)
else:
    y = y_series.values.astype(int)
    class_names = np.unique(y).tolist()

n_classes = len(np.unique(y))
print("Detected classes:", class_names, " n_classes=", n_classes)

X = X_df.values.astype(np.float32)
n_samples, n_features = X.shape

# Try to infer CHANNELS and TIMESTEPS if not provided
if CHANNELS is None and TIMESTEPS is None:
    # try to see if n_features is divisible by a plausible channel count
    # common EEG channels: 19, 21, 23, 25, or single-channel
    plausible_channels = [1, 2, 4, 8, 16, 19, 21, 23, 32]
    found = False
    for ch in plausible_channels:
        if n_features % ch == 0 and n_features // ch >= 16:
            CHANNELS = ch
            TIMESTEPS = n_features // ch
            found = True
            break
    if not found:
        # fallback to treat as single-channel time series
        CHANNELS = 1
        TIMESTEPS = n_features
elif CHANNELS is None:
    CHANNELS = 1
    TIMESTEPS = n_features // CHANNELS
elif TIMESTEPS is None:
    TIMESTEPS = n_features // CHANNELS

# final check
if CHANNELS * TIMESTEPS != n_features:
    # fallback: try single-channel
    print(f"WARNING: CHANNELS({CHANNELS}) * TIMESTEPS({TIMESTEPS}) != n_features({n_features}). Using single-channel flatten.")
    CHANNELS = 1
    TIMESTEPS = n_features

print(f"Reshaping X -> (samples, timesteps, channels) = ({n_samples}, {TIMESTEPS}, {CHANNELS})")
X = X.reshape((n_samples, TIMESTEPS, CHANNELS))

# Standardize per-feature (flattened), fit on training later to avoid leakage
# We'll standardize across timesteps per channel using StandardScaler on flattened features
X_flat = X.reshape((n_samples, TIMESTEPS * CHANNELS))
scaler = StandardScaler()
X_flat = scaler.fit_transform(X_flat)
X = X_flat.reshape((n_samples, TIMESTEPS, CHANNELS))

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=SEED, stratify=y)
print("Train:", X_train.shape, y_train.shape, " Test:", X_test.shape, y_test.shape)

# ----------------- Model building blocks -----------------
def conv_block(inputs, filters, kernel_size, pool_size=2, dropout=0.2):
    x = layers.Conv1D(filters, kernel_size, padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size)(x)
    x = layers.Dropout(dropout)(x)
    return x

def se_block(inputs, ratio=8):
    # Squeeze-and-Excitation channel attention (simple)
    filters = int(inputs.shape[-1])
    se = layers.GlobalAveragePooling1D()(inputs)
    se = layers.Dense(filters // ratio, activation='relu')(se)
    se = layers.Dense(filters, activation='sigmoid')(se)
    se = layers.Reshape((1, filters))(se)
    return layers.multiply([inputs, se])

# CNN-BiLSTM branch
def build_cnn_bilstm_branch(input_shape, name="cnn_bilstm_branch"):
    inp = layers.Input(shape=input_shape)
    x = conv_block(inp, filters=32, kernel_size=7, pool_size=2)
    x = conv_block(x, filters=64, kernel_size=5, pool_size=2)
    x = conv_block(x, filters=128, kernel_size=3, pool_size=2)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=False))(x)  # output vector
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

# MASF-like multi-scale attention branch (simplified)
def build_masf_branch(input_shape, name="masf_branch"):
    inp = layers.Input(shape=input_shape)
    # Multi-scale conv branches
    b1 = layers.Conv1D(32, kernel_size=3, padding='same', activation='relu')(inp)
    b1 = layers.BatchNormalization()(b1)
    b1 = layers.MaxPooling1D(2)(b1)

    b2 = layers.Conv1D(32, kernel_size=5, padding='same', activation='relu')(inp)
    b2 = layers.BatchNormalization()(b2)
    b2 = layers.MaxPooling1D(2)(b2)

    b3 = layers.Conv1D(32, kernel_size=11, padding='same', activation='relu')(inp)
    b3 = layers.BatchNormalization()(b3)
    b3 = layers.MaxPooling1D(2)(b3)

    x = layers.Concatenate(axis=-1)([b1, b2, b3])
    x = se_block(x)  # attention
    x = layers.Conv1D(128, kernel_size=3, padding='same', activation='relu')(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    return models.Model(inp, x, name=name)

# Fusion model: feature-level fusion
def build_hybrid_model(input_shape, n_classes, freeze_branches=False):
    # Build branches
    branch1 = build_cnn_bilstm_branch(input_shape)
    branch2 = build_masf_branch(input_shape)
    # Optionally freeze pretrained branch layers (if you want to load weights)
    if freeze_branches:
        branch1.trainable = False
        branch2.trainable = False

    # Inputs
    inp = layers.Input(shape=input_shape)
    f1 = branch1(inp)
    f2 = branch2(inp)

    # Concatenate features
    fused = layers.Concatenate()([f1, f2])
    # Optional attention / gating on fused vector
    g = layers.Dense(256, activation='relu')(fused)
    g = layers.Dropout(0.4)(g)
    g = layers.Dense(128, activation='relu')(g)
    g = layers.Dropout(0.3)(g)
    out = layers.Dense(n_classes, activation='softmax')(g)

    model = models.Model(inp, out, name="Hybrid_CNNBiLSTM_MASFNet")
    return model

# Build the model
input_shape = (TIMESTEPS, CHANNELS)
model = build_hybrid_model(input_shape, n_classes, freeze_branches=False)
model.compile(optimizer=optimizers.Adam(LR),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# ----------------- Training -----------------
# Callbacks
es = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1)
rlp = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-6)
ckpt_path = "best_hybrid_model.h5"
mc = callbacks.ModelCheckpoint(ckpt_path, monitor='val_loss', save_best_only=True, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_split=VAL_SPLIT,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[es, rlp, mc],
    verbose=2,
    shuffle=True
)

# load best weights
if os.path.exists(ckpt_path):
    model.load_weights(ckpt_path)

# ----------------- Evaluation -----------------
y_pred_probs = model.predict(X_test, batch_size=BATCH_SIZE)
y_pred = np.argmax(y_pred_probs, axis=1)
test_acc = accuracy_score(y_test, y_pred)
print(f"\nHybrid model test accuracy: {test_acc*100:.2f}%\n")

print("Classification report:")
print(classification_report(y_test, y_pred, target_names=[str(c) for c in class_names], digits=4))

print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

# Plot training curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title('Loss')

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('Accuracy')

plt.show()

# ----------------- Decision-level fusion (optional quick eval) -----------------
# If you already have separately trained models for CNN-BiLSTM and MASFNet saved,
# you can load them and average probabilities to get a decision-level ensemble.
# Example (sketch):
#   probs1 = model1.predict(X_test)
#   probs2 = model2.predict(X_test)
#   final_probs = 0.6*probs1 + 0.4*probs2
#   y_ensemble = np.argmax(final_probs, axis=1)
#   print(accuracy_score(y_test, y_ensemble))
#
# If you want this code for real pretrained branch models, tell me and I will add it.
````
